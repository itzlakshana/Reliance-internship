```python
from statsmodels.tsa.arima.model import ARIMA
X = data3.values
size = int(len(X) * 0.52)
train, test = X[0:size], X[size:len(X)]
train_lst = [x for x in train]
model = ARIMA(train_lst, order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
X = data3.values
size = int(len(X) * 0.41)
train, test = X[0:size], X[size:len(X)]
train_lst = [x for x in train]
predictions = list()
# walk-forward validation
for t in range(len(test)):
    model = ARIMA(train_lst, order=(1,0,1))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    train_lst.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
# evaluate forecasts
rmse = math.sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
```

```python
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

```python
# simple exponential smoothing
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
X = data3.values
size = int(len(X) * 0.44)
train, test = X[0:size], X[size:len(X)]
train_lst = [x for x in train]
ses_model = SimpleExpSmoothing(train_lst).fit(smoothing_level=0.3)
y_pred = ses_model.forecast(51)
math.sqrt(mean_absolute_error(test, y_pred))
```

```python
#plt.plot(train)
plt.plot(test, color='red')
plt.plot(y_pred, color='green')
plt.show()
```

```python
#Double Exponential Smoothing or holt model
from statsmodels.tsa.holtwinters import ExponentialSmoothing
X = data3.values
size = int(len(X) * 0.40)
train, test = X[0:size], X[size:len(X)]
train_lst = [x for x in train]
# Additive model
model_add = ExponentialSmoothing(train_lst, trend='add')
results_add = model_add.fit()
predictions_add = results_add.forecast(steps=36)
# Multiplicative model
model_mul = ExponentialSmoothing(train_lst, trend='mul')
results_mul = model_mul.fit()
predictions_mul = results_mul.forecast(steps=36)
# Evaluate
rmse_add = math.sqrt(mean_squared_error(train_lst, predictions_add))
rmse_mul = math.sqrt(mean_squared_error(train_lst, predictions_mul))
```

```python
plt.title(f'Double Exponential Smoothing predictions\nRMSE (add) = {np.round(rmse_add, 2)} | RMSE (mul) = {np.round(rmse_mul, 2)}', size=18)
plt.plot(train_lst, label='Training data')
plt.plot(test, color='gray', label='Testing data')
plt.plot(predictions_add, color='orange', label='Predictions Additive')
plt.plot(predictions_mul, color='red', label='Predictions Multiplicative')
plt.legend();
```

```python
#Triple Exponential Smoothing or Holt-Winters
from statsmodels.tsa.holtwinters import ExponentialSmoothing
X = data3.values
size = int(len(X) * 0.42)
train, test = X[0:size], X[size:len(X)]
train_lst = [x for x in train]
# Multiplicative trend / Additive seasonality model
model_mul_add = ExponentialSmoothing(train, trend='mul', seasonal='add', seasonal_periods=12)
results_mul_add = model_mul_add.fit()
predictions_mul_add = results_mul_add.forecast(steps=38)

# Multiplicative trend / Multiplicative seasonality model
model_mul_mul = ExponentialSmoothing(test, trend='mul', seasonal='mul', seasonal_periods=12)
results_mul_mul = model_mul_mul.fit()
predictions_mul_mul = results_mul_mul.forecast(steps=53)

# Evaluate
rmse_mul_add = math.sqrt(mean_squared_error(train, predictions_mul_add))
rmse_mul_mul = math.sqrt(mean_squared_error(test, predictions_mul_mul))
```

```python
plt.title(f'Triple Exponential Smoothing predictions\nRMSE (mul, add) = {np.round(rmse_mul_add, 2)} | RMSE (mul, mul) = {np.round(rmse_mul_mul, 2)}', size=20)
plt.plot(train, label='Training data')
plt.plot(test, color='gray', label='Testing data')
plt.plot(predictions_mul_add, color='orange', label='Predictions Multiplicative trend, Additive seasonality')
plt.plot(predictions_mul_mul, color='red', label='Predictions Multiplicative trend, Multiplicative seasonality')
plt.legend()
```
